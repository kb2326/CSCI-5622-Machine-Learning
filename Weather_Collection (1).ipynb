```python
import pandas as pd

# Define the file paths and corresponding city names
city_files = {
    'Denver': r"C:\Users\karth\Downloads\Denver.csv",
    'Greely': r"C:\Users\karth\Downloads\Greely.csv",
    'Aspen': r"C:\Users\karth\Downloads\Aspen.csv",
    'Durango': r"C:\Users\karth\Downloads\Durango.csv",
    'GrandJunction': r"C:\Users\karth\Downloads\GrandJunction.csv",
    'Pueblo': r"C:\Users\karth\Downloads\Pueblo.csv",
    'FortCollins': r"C:\Users\karth\Downloads\FortCollins.csv",
    'ColoradoSprings': r"C:\Users\karth\Downloads\Colorado Springs.csv",
    'Boulder': r"C:\Users\karth\Downloads\Boulder.csv"
}

# Initialize an empty DataFrame to hold all data
all_data = pd.DataFrame()

# Loop through each city and file, read the data, add the city name as a column, and append to the main DataFrame
for city,file_path in city_files.items():
    temp_df = pd.read_csv(file_path)
    temp_df['City'] = city  # Add city name as a new column
    all_data = pd.concat([all_data, temp_df], ignore_index=True)
```

```python
all_data[['YEAR', 'MO', 'DY', 'HR']] = all_data[['YEAR', 'MO', 'DY', 'HR']].fillna(0).astype(int)
```

```python
all_data['timestamp'] = pd.to_datetime(
    all_data['YEAR'].astype(str) + '-' +
    all_data['MO'].astype(str).str.zfill(2) + '-' +
    all_data['DY'].astype(str).str.zfill(2) + ' ' +
    all_data['HR'].astype(str).str.zfill(2) + ':00:00'
)

# Set the timestamp as the index
all_data.set_index('timestamp', inplace=True)
```

```python
weather_columns = ['T2M', 'T2MDEW', 'T2MWET', 'QV2M', 'RH2M', 'PRECTOTCORR', 'PS', 'WS2M', 'WD2M', 'WS10M']

# Group by the timestamp index and calculate the mean for each weather variable
average_weather = all_data[weather_columns].groupby(all_data.index).mean()

# Reset index to turn the datetime index back into a column, if you need it for further analysis
average_weather.reset_index(inplace=True)

# Print the resulting DataFrame to verify
average_weather.head()
```
